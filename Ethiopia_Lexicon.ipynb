```
terms_all = terms1.append([terms2, terms3, terms4, terms5, terms6, terms7, 
                           terms8, terms9,terms10, terms11, terms12])
terms_all.head()
```

```
#Dropiing all rows will null values (translation not available)
terms_final = terms_all.dropna(how='any',axis=0)
terms_final.shape[0]
```
